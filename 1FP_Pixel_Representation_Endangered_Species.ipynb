```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
img_name = "vaquita.jpg"
species = 30

img = cv2.imread(img_name);
cv2_imshow(img);
```

```
height, width = img.shape[:2]
pixel = int(np.floor(np.sqrt(species)))      #it will have the same amount of pixels in both directions
print(pixel)

x = int(np.floor(height/pixel));          #height of each pixel
y = int(np.floor(width/pixel));           #width of each pixel
print(x,y)

img2 = img[0:pixel*x, 0:pixel*y,:].copy()
for i in range(pixel):
    for j in range(pixel):
        for k in range(3):
            v = int(np.floor(img[i*x:(i+1)*x, j*y:(j+1)*y, k].sum()/(x*y)))     #average value of the channel
            img2[i*x:(i+1)*x, j*y:(j+1)*y, k] = v

cv2_imshow(img2)
```
